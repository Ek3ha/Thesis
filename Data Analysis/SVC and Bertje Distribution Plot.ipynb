```python
import pandas as pd
data = pd.read_csv("Bertje-Narrative-Distribution.csv")
```

```python
bertje_distribution = data["0"].value_counts()
```

```python
data1 = pd.read_csv("svc-Narrative-Distribution.csv")
```

```python
data_svc = data1["0"].value_counts()
```

```python


all_labels = sorted(set(data_svc.index).union(set(bertje_distribution.index)))

bertje_distribution = bertje_distribution.reindex(all_labels,fill_value=0)
data_svc = data_svc.reindex(all_labels,fill_value=0)

len(bertje_distribution), len(data_svc)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the distribution
colours = sns.color_palette("pastel",12)
plt.figure(figsize=(8, 25))

fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.3
index = np.arange(len(data_svc))

#bar1 = data_label_counts.plot(kind='bar' ,width=bar_width, label='Bertje', color=colours[0],edgecolor="black")
#bar2 = data_svc.plot(kind='bar' ,width=bar_width, label='SVC', color=colours[0],edgecolor="black")
bar1 = ax.bar(index, data_svc, bar_width, label='SVC', color=colours[0],edgecolor='black')
bar2 = ax.bar(index + bar_width, bertje_distribution, bar_width, label='Bertje',color=colours[1],edgecolor='black')

plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.title('Distribution of Narrative classification on the housing dataset')

ax.set_xticks(index + bar_width )
ax.set_xticklabels(data_svc.index, rotation='vertical')
ax.legend()


plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```
